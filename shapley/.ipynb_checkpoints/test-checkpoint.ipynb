```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.special import expit
from scipy.io import loadmat
import shap
import seaborn as sns
```

```python

def Actor_NL_gamma_bah(X):
    X=X.T

    # Activity function for value function
    Neuron_Num_a = 20
    op = v_actor.T@ X
    DS = 1 / (1 + np.exp(-op))

    return DS


def model_(X):
    u = W_actor.T @ Actor_NL_gamma_bah(X)
    return np.array(u).reshape(-1,1)
```

```python
mat_data = loadmat('New_MNN_SVD.mat')
Wa = mat_data['W_actor']
v_actor = mat_data['Va']
df = mat_data['X'].T
y = mat_data['u']
mat_data.clear()
W_actor = Wa[:,:,-1]
X = pd.DataFrame(np.array(df),columns=['e1','e2','r1','r2','bias'])
y = pd.DataFrame(np.array(y).reshape(-1,1), columns=['Control_input'])
X.head()
```

```python
background_data = shap.utils.sample(X, nsamples=100 ,random_state=0)
background_data.head()
```

```python
#model_(background_data)
```

```python
# Create a SHAP explainer
explainer = shap.KernelExplainer(model_,background_data)
explainer
```

```python
# Compute SHAP values
shap_ = explainer.shap_values(X)
```

```python
shap_values=shap_[0]
```

```python
X.iloc[1000,:]
```
